```python
import cv2
import pyttsx3
from ultralytics import YOLO
import time

# Initialize TTS engine
engine = pyttsx3.init()
engine.setProperty('rate', 150)  # Adjust speech rate

# Load YOLOv8 model (Nano model is fastest)
model = YOLO('yolov8n.pt')  # Automatically downloads on first run

# Start video capture
cap = cv2.VideoCapture(0)  # Use 0 for built-in webcam

# Track objects already announced
spoken_objects = set()

print("🔵 VisionMate is running... Press 'q' to quit.")
last_reset_time = time.time()
reset_interval = 10  # in seconds (you can change this)

while True:
    # Reset spoken_objects every X seconds
    current_time = time.time()
    if current_time - last_reset_time > reset_interval:
        spoken_objects.clear()
        last_reset_time = current_time

    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO object detection
    results = model(frame, stream=True)
    detected_labels = []

    for r in results:
        for box in r.boxes:
            cls_id = int(box.cls[0])
            label = model.names[cls_id]

            # Get bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cx = (x1 + x2) // 2  # Center x-coordinate
            
            # Determine direction
            frame_width = frame.shape[1]
            if cx < frame_width * 0.33:
                direction = "on the left"
            elif cx > frame_width * 0.66:
                direction = "on the right"
            else:
                direction = "in the center"
            
            # Create a label with direction
            full_label = f"{label} {direction}"
            detected_labels.append(full_label)
            
            # Draw the box and label
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, full_label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Speak out new labels
    for label in detected_labels:
        if label not in spoken_objects:
            engine.say(f"{label} ahead")
            engine.runAndWait()
            spoken_objects.add(label)

    # Show output
    cv2.imshow("VisionMate - AI Assistant", frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
